```python
import logging
import time
from functools import wraps

# -----------------------------
# Logging Configuration
# -----------------------------

logging.basicConfig(
    level=logging.INFO,
    filename="api_logs.log",
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logger = logging.getLogger("API_MIDDLEWARE")

# -----------------------------
# Timing Decorator
# -----------------------------

def timer(func):

    @wraps(func)
    def wrapper(*args, **kwargs):

        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()

        logger.info(f"{func.__name__} executed in {end-start:.4f} seconds")

        return result

    return wrapper


# -----------------------------
# Logging Decorator
# -----------------------------

def log_io(func):

    @wraps(func)
    def wrapper(*args, **kwargs):

        logger.info(f"Function {func.__name__} called with args {args}")

        result = func(*args, **kwargs)

        logger.info(f"Function {func.__name__} returned {result}")

        return result

    return wrapper


# -----------------------------
# Access Control Decorator
# -----------------------------

def require_role(role):

    def decorator(func):

        @wraps(func)
        def wrapper(user_role, *args, **kwargs):

            if user_role != role:

                logger.error("Access Denied")

                return "Access Denied"

            return func(user_role, *args, **kwargs)

        return wrapper

    return decorator


# -----------------------------
# Example API Function
# -----------------------------

@timer
@log_io
@require_role("admin")

def delete_user(user_role, username):

    time.sleep(1)

    return f"User {username} deleted"


# -----------------------------
# Testing
# -----------------------------

print(delete_user("admin", "Ali"))

print(delete_user("guest", "Sara"))
```
