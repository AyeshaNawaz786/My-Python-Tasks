```python
# ==========================================
# TASK 3 
# ==========================================

import os
from datetime import datetime, timedelta
import pandas as pd

# ------------------------------
# Configuration
# ------------------------------

PROJECT_FOLDER = "student_records"
AUDIT_REPORT = "audit_report.csv"
PENALTY_REPORT = "penalty_report.csv"

audit_data = []
penalty_data = []

# ------------------------------
# Show current directory
# ------------------------------

print("📂 Current Working Directory:")
print(os.getcwd())
print("-" * 50)

# ------------------------------
# Check folder
# ------------------------------

if not os.path.exists(PROJECT_FOLDER):
    print("❌ Folder not found.")
else:

    files = os.listdir(PROJECT_FOLDER)

    print(f"📁 Files Found: {len(files)}\n")

    now = datetime.now()
    threshold = now - timedelta(days=30)

    # ------------------------------
    # Scan files
    # ------------------------------

    for file in files:

        file_path = os.path.join(PROJECT_FOLDER, file)

        if os.path.isfile(file_path):

            stats = os.stat(file_path)

            size_kb = round(stats.st_size / 1024, 2)
            modified_time = datetime.fromtimestamp(stats.st_mtime)

            days_old = (now - modified_time).days

            old_file = days_old > 30

            audit_data.append({
                "file_name": file,
                "size_kb": size_kb,
                "last_modified": modified_time.strftime("%Y-%m-%d"),
                "days_old": days_old,
                "old_file": old_file
            })

            # penalty rule
            if old_file:
                fine = round(size_kb * 2, 2)

                penalty_data.append({
                    "file_name": file,
                    "days_old": days_old,
                    "fine_amount": fine
                })

# ------------------------------
# Save audit report
# ------------------------------

audit_df = pd.DataFrame(audit_data)
audit_df.to_csv(AUDIT_REPORT, index=False)

# ------------------------------
# Save penalty report
# ------------------------------

penalty_df = pd.DataFrame(penalty_data)
penalty_df.to_csv(PENALTY_REPORT, index=False)

# ------------------------------
# Summary
# ------------------------------

print("✅ Audit Completed Successfully\n")

print("📊 Summary")
print("Total Files Scanned:", len(audit_df))
print("Old Files:", len(penalty_df))

print("\n📄 Reports Generated:")
print(AUDIT_REPORT)
print(PENALTY_REPORT)
```
