```python
class BankAccount:

    bank_name = "Ayesha Bank"

    def __init__(self, owner, balance=0):

        self.owner = owner
        self.__balance = balance

    # deposit method
    def deposit(self, amount):

        if amount > 0:
            self.__balance += amount
            print(f"{amount} deposited successfully")

        else:
            print("Invalid deposit amount")

    # withdraw method
    def withdraw(self, amount):

        if amount > self.__balance:
            print("Insufficient balance")

        elif amount <= 0:
            print("Invalid withdrawal amount")

        else:
            self.__balance -= amount
            print(f"{amount} withdrawn successfully")

    # property getter
    @property
    def balance(self):

        return self.__balance

    # class method
    @classmethod
    def bank_info(cls):

        print("Welcome to", cls.bank_name)

    # static method
    @staticmethod
    def bank_rules():

        print("Rule: Balance cannot be negative")

    # readable output
    def __str__(self):

        return f"Account Owner: {self.owner} | Balance: {self.__balance}"


# main program
if __name__ == "__main__":

    BankAccount.bank_info()

    account1 = BankAccount("Ali", 5000)

    account1.deposit(2000)

    account1.withdraw(1500)

    print("\nAccount Summary")
    print(account1)

    BankAccount.bank_rules()
```
