```python
# ==========================================
# TASK 4 
# ==========================================

products = [
    {"name":"Laptop","category":"Electronics","price":"900"},
    {"name":"Phone","category":"Electronics","price":"600"},
    {"name":"Tablet","category":"Electronics","price":None},
    {"name":"Headphones","category":"Electronics","price":"150"},
    {"name":"TV","category":"Electronics","price":"1200"},
    {"name":"Camera","category":"Electronics","price":"800"},
    
    {"name":"Shirt","category":"Clothing","price":"40"},
    {"name":"Jeans","category":"Clothing","price":"60"},
    {"name":"Jacket","category":"Clothing","price":"120"},
    {"name":"Shoes","category":"Clothing","price":"90"},
    {"name":"Cap","category":"Clothing","price":None},
    
    {"name":"Laptop","category":"Electronics","price":"900"} # duplicate
]

# ----------------------------
# Step 1 Clean Data
# ----------------------------

clean_products = []

seen = set()

for p in products:

    name = p["name"]
    price = p["price"]

    if price is None:
        continue

    price = float(price)

    if name in seen:
        continue

    seen.add(name)

    clean_products.append({
        "name":name,
        "category":p["category"],
        "price":price
    })

# ----------------------------
# Step 2 Group by Category
# ----------------------------

categories = {}

for p in clean_products:
    cat = p["category"]

    if cat not in categories:
        categories[cat] = []

    categories[cat].append(p)

# ----------------------------
# Step 3 Find Top 5 Expensive
# ----------------------------

summary = {}

for cat,items in categories.items():

    top_items = sorted(
        items,
        key=lambda x: x["price"],
        reverse=True
    )[:5]

    summary[cat] = top_items

# ----------------------------
# Step 4 Print Result
# ----------------------------

print("Inventory Summary\n")

for cat,items in summary.items():

    print(cat)
    
    for p in items:
        print(f"  {p['name']} - ${p['price']}")
    
    print()
```

```python
# ==============================================
# Task 4 – Part 2
# HR Employee Ranker
# ==============================================

# Large Employee Dataset (50+ records)

employees = [
{"name":"Ali","score":88,"experience":5},
{"name":"Sara","score":92,"experience":3},
{"name":"Ahmed","score":92,"experience":6},
{"name":"Zain","score":75,"experience":4},
{"name":"Hina","score":88,"experience":7},
{"name":"Bilal","score":92,"experience":6},
{"name":"Ayesha","score":85,"experience":2},
{"name":"Hamza","score":79,"experience":4},
{"name":"Fatima","score":91,"experience":5},
{"name":"Usman","score":87,"experience":3},
{"name":"Kiran","score":90,"experience":6},
{"name":"Saad","score":83,"experience":2},
{"name":"Nida","score":89,"experience":4},
{"name":"Imran","score":94,"experience":8},
{"name":"Sana","score":86,"experience":3},
{"name":"Farhan","score":78,"experience":2},
{"name":"Mahnoor","score":91,"experience":4},
{"name":"Danish","score":84,"experience":3},
{"name":"Hassan","score":80,"experience":5},
{"name":"Tariq","score":88,"experience":6},
{"name":"Adnan","score":82,"experience":3},
{"name":"Arham","score":90,"experience":4},
{"name":"Basit","score":76,"experience":2},
{"name":"Dua","score":87,"experience":4},
{"name":"Eman","score":89,"experience":5},
{"name":"Fahad","score":91,"experience":7},
{"name":"Ghulam","score":81,"experience":3},
{"name":"Hafsa","score":88,"experience":6},
{"name":"Iqra","score":86,"experience":4},
{"name":"Junaid","score":93,"experience":5},
{"name":"Kamran","score":84,"experience":2},
{"name":"Laiba","score":90,"experience":3},
{"name":"Muneeb","score":82,"experience":4},
{"name":"Noman","score":79,"experience":3},
{"name":"Omer","score":91,"experience":6},
{"name":"Parveen","score":85,"experience":5},
{"name":"Qasim","score":83,"experience":3},
{"name":"Rida","score":88,"experience":4},
{"name":"Shahzaib","score":92,"experience":5},
{"name":"Talha","score":80,"experience":4},
{"name":"Urooj","score":87,"experience":6},
{"name":"Waqas","score":89,"experience":5},
{"name":"Yasir","score":84,"experience":4},
{"name":"Zara","score":90,"experience":6},
{"name":"Adeel","score":86,"experience":3},
{"name":"Bisma","score":88,"experience":5},
{"name":"Chetan","score":77,"experience":2},
{"name":"Dawood","score":83,"experience":4},
{"name":"Ehsan","score":91,"experience":7},
{"name":"Faiza","score":85,"experience":3},
{"name":"Gohar","score":82,"experience":4},
{"name":"Haris","score":89,"experience":6}
]

# ------------------------------------------------
# Step 1: Sort Employees
# Rule:
# Score (High → Low)
# Experience (High → Low)
# Name (Alphabetical)
# ------------------------------------------------

sorted_employees = sorted(
    employees,
    key=lambda e: (-e["score"], -e["experience"], e["name"])
)

print("🏆 Employee Ranking\n")

for rank, emp in enumerate(sorted_employees, start=1):
    print(f"{rank}. {emp['name']} | Score: {emp['score']} | Experience: {emp['experience']} years")


# ------------------------------------------------
# Step 2: Prepare names for binary search
# ------------------------------------------------

names = sorted([emp["name"].lower() for emp in sorted_employees])


# ------------------------------------------------
# Step 3: Binary Search
# ------------------------------------------------

def binary_search(data, target):

    low = 0
    high = len(data) - 1

    while low <= high:

        mid = (low + high) // 2

        if data[mid] == target:
            return mid

        elif data[mid] < target:
            low = mid + 1

        else:
            high = mid - 1

    return -1


# ------------------------------------------------
# Step 4: Search Employee
# ------------------------------------------------

search_name = input("\nEnter employee name to search: ").lower()

index = binary_search(names, search_name)

if index != -1:

    for emp in sorted_employees:
        if emp["name"].lower() == search_name:

            print("\n✅ Employee Found")
            print(f"Name: {emp['name']}")
            print(f"Score: {emp['score']}")
            print(f"Experience: {emp['experience']} years")

else:
    print("❌ Employee not found")
```
