```python
from collections import defaultdict

# Step 1: create log file automatically
log_data = """INFO - Server started
INFO - User login
ERROR - Database connection failed
WARNING - Disk usage high
CRITICAL - Memory overflow
ERROR - Timeout error
ERROR - Database connection failed
"""

with open("server.log", "w") as f:
    f.write(log_data)


# Step 2: generator function
def error_streamer(file_path):

    with open(file_path, "r") as file:
        for line in file:
            if "ERROR" in line or "CRITICAL" in line:
                yield line.strip()


# Step 3: process errors
def process_errors(file_path):

    error_counts = defaultdict(int)

    for error_line in error_streamer(file_path):

        parts = error_line.split(" - ")

        if len(parts) > 1:
            error_type = parts[1]
        else:
            error_type = "Unknown Error"

        error_counts[error_type] += 1

        print("LIVE:", error_type, "->", error_counts[error_type])

    return error_counts


# Step 4: main program
def main():

    summary = process_errors("server.log")

    print("\nFinal Error Summary\n")

    for error, count in summary.items():
        print(error, ":", count)


if __name__ == "__main__":
    main()
```
