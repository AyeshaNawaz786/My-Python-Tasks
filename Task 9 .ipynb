```python
from abc import ABC, abstractmethod
from datetime import datetime

class Book:

    def __init__(self, title):
        self.title = title
        self.available = True
        self.borrow_date = None

    def __str__(self):
        status = "Available" if self.available else "Borrowed"
        return f"{self.title} ({status})"


class Member(ABC):

    def __init__(self, name, limit):
        self.name = name
        self.limit = limit
        self.borrowed_books = []

    def borrow_book(self, book):

        if len(self.borrowed_books) >= self.limit:
            print(f"{self.name} cannot borrow more books (limit reached)\n")
            return

        if book.available:

            book.available = False
            book.borrow_date = datetime.now()

            self.borrowed_books.append(book)

            print(f"{self.name} borrowed '{book.title}'\n")

        else:
            print(f"{book.title} is not available\n")

    def return_book(self, book):

        if book in self.borrowed_books:

            days = (datetime.now() - book.borrow_date).days
            fine = self.calculate_fine(days)

            book.available = True
            self.borrowed_books.remove(book)

            print(f"{self.name} returned '{book.title}'")

            if fine > 0:
                print(f"Late Fine: {fine}")

            print()

        else:
            print("Book was not borrowed by this member\n")

    def show_borrowed_books(self):

        print(f"{self.name}'s Borrowed Books:")

        if not self.borrowed_books:
            print("No borrowed books\n")

        else:
            for book in self.borrowed_books:
                print("-", book.title)
            print()

    @abstractmethod
    def calculate_fine(self, days):
        pass


class RegularMember(Member):

    def __init__(self, name):
        super().__init__(name, limit=2)

    def calculate_fine(self, days):

        if days > 7:
            return (days - 7) * 5
        return 0


class PremiumMember(Member):

    def __init__(self, name):
        super().__init__(name, limit=5)

    def calculate_fine(self, days):

        if days > 14:
            return (days - 14) * 2
        return 0


class Library:

    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_books(self):

        print("\nLibrary Inventory:\n")

        for book in self.books:
            print(book)

        print()

    def __len__(self):
        return len(self.books)


if __name__ == "__main__":

    library = Library()

    library.add_book(Book("Python Programming"))
    library.add_book(Book("Data Structures"))
    library.add_book(Book("Machine Learning"))
    library.add_book(Book("Database Systems"))
    library.add_book(Book("Operating Systems"))

    library.show_books()

    member1 = RegularMember("Ali")
    member2 = PremiumMember("Sara")

    member1.borrow_book(library.books[0])
    member1.borrow_book(library.books[1])

    member2.borrow_book(library.books[2])

    member1.show_borrowed_books()
    member2.show_borrowed_books()

    member1.return_book(library.books[0])

    print("Total books in library:", len(library))
```
