```python
# Task 10

import asyncio
import time
from concurrent.futures import ThreadPoolExecutor


# -------- THREAD TASKS --------
def generate_report():
    print("Generating report...")
    time.sleep(2)
    return "Report Generated"


def save_file():
    print("Saving file...")
    time.sleep(2)
    return "File Saved Successfully"


# -------- ASYNC TASKS --------
async def fetch_students():
    print("Fetching students data...")
    await asyncio.sleep(2)
    return {"students": 150}


async def fetch_courses():
    print("Fetching courses data...")
    await asyncio.sleep(2)
    return {"courses": 6}


async def process_data(student_data, course_data):
    print("Processing data...")
    await asyncio.sleep(2)

    combined = {
        "students": student_data["students"],
        "courses": course_data["courses"],
        "status": "Processed"
    }

    return combined


# -------- MAIN FUNCTION --------
async def main():

    print("System Starting...\n")

    # Run async tasks concurrently
    student_data, course_data = await asyncio.gather(
        fetch_students(),
        fetch_courses()
    )

    processed_data = await process_data(student_data, course_data)

    # Run threaded tasks
    with ThreadPoolExecutor() as executor:

        report_future = executor.submit(generate_report)
        file_future = executor.submit(save_file)

        report_result = report_future.result()
        file_result = file_future.result()

    print("\nFinal Results\n")

    print("Processed Data :", processed_data)
    print("Report Status :", report_result)
    print("File Status :", file_result)


# Run program
await main()
```
