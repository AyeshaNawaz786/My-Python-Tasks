```python
# ==============================================
# Day 2
# ==============================================

import os
import math
import pandas as pd

INPUT_FOLDER = "student_records"
OUTPUT_FILE = "clean_results.csv"

cleaned_results = []
processed_students = set()

# Check folder
if not os.path.exists(INPUT_FOLDER):
    print("❌ Folder does not exist.")
else:
    files = os.listdir(INPUT_FOLDER)
    print(f"📂 Found {len(files)} files.\n")

    for file_name in files:
        if file_name.endswith(".xlsx"):

            file_path = os.path.join(INPUT_FOLDER, file_name)
            print(f"🔄 Processing: {file_name}")

            try:
                df = pd.read_excel(file_path)

                for index, row in df.iterrows():
                    try:
                        name = row.get("name")
                        marks = row.get("marks")

                        # Validate name
                        if pd.isna(name) or str(name).strip() == "":
                            continue

                        # Validate marks
                        if pd.isna(marks):
                            continue

                        marks = float(marks)

                        # Check NaN
                        if math.isnan(marks):
                            continue

                        # Range validation
                        if marks < 0 or marks > 100:
                            continue

                        # Remove duplicates
                        if str(name).strip() in processed_students:
                            continue

                        processed_students.add(str(name).strip())

                        status = "Pass" if marks >= 40 else "Fail"

                        cleaned_results.append({
                            "name": str(name).strip(),
                            "marks": marks,
                            "status": status,
                            "source_file": file_name
                        })

                    except (ValueError, TypeError):
                        continue

            except Exception as e:
                print(f"⚠ Error reading {file_name}: {e}")

# Save clean output
try:
    final_df = pd.DataFrame(cleaned_results)
    final_df.to_csv(OUTPUT_FILE, index=False)

    print("\n✅ Clean results file generated successfully!")
    print(f"📊 Total unique valid students: {len(final_df)}")

except Exception as e:
    print("❌ Error writing output file:", e)
```
