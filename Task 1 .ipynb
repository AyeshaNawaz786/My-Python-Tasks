```python
# ==========================================
# Product Catalogue Cleaner
# Internship Evaluation - Day 1
# ==========================================

# Raw messy supplier data
raw_products = [
    {"name": " Laptop ", "price": "50000", "category": " Electronics "},
    {"name": "Mouse", "price": 1500, "category": "Electronics"},
    {"name": "Laptop", "price": "50000", "category": "Electronics"},  # duplicate
    {"name": "Shirt", "price": "2000", "category": "Clothing"},
    {"name": "Jeans", "price": None, "category": "Clothing"},  # missing price
    {"name": "Headphones", "price": "abc", "category": "Electronics"},  # invalid price
    {"name": "Shoes", "price": "3,000", "category": " clothing "}
]

cleaned_products = []
seen_products = set()

for product in raw_products:
    
    # Safely get values
    name = product.get("name")
    price = product.get("price")
    category = product.get("category")
    
    # Skip if any important field missing
    if not name or not category or price is None:
        continue
    
    # Clean strings
    name = name.strip()
    category = category.strip().title()   # Normalize case
    
    # Remove comma from price if exists
    if isinstance(price, str):
        price = price.replace(",", "").strip()
    
    # Safe float conversion
    try:
        price = float(price)
    except (ValueError, TypeError):
        continue
    
    # Duplicate check based on name + category
    product_key = (name.lower(), category.lower())
    
    if product_key in seen_products:
        continue
    
    seen_products.add(product_key)
    
    cleaned_products.append({
        "name": name,
        "price": price,
        "category": category
    })

# Group by category
grouped_products = {}

for product in cleaned_products:
    category = product["category"]
    
    if category not in grouped_products:
        grouped_products[category] = []
    
    grouped_products[category].append({
        "name": product["name"],
        "price": product["price"]
    })

# Print neat summary
print("==== CLEANED PRODUCT SUMMARY ====\n")

for category, items in grouped_products.items():
    print(f"Category: {category}")
    for item in items:
        print(f"  - {item['name']} : Rs {item['price']}")
    print()
```
